# Merging Two data frames inot one

# keep only those columns 'features' which are important, rest of them are ignored

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.shape
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.head(2)
```

there are some similar words like [love,loved,loving] for elimiting this we will use technique
which is called stemming, Stemming is a natural language processing technique and nltk library, that lowers inflection in words to their root forms, when applying stemming we will have [love,love,love]

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

dong vectorization to change each tag into vector, and then the ligc will be like near vector will be the recommendeton for that movie, maxt_feature is for how many words you want to convert to vectro, and stop_words is used for to remove stop words which is conbributing the meaning of sentences like(is,or,and,of for...)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

using cv object and pass tag to convert to vectro, it will convert it to metrix we convert it numypy arre explicity

```python
vector = cv.fit_transform(new['tags']).toarray()
```

now we will have 4806 rows and 5000 columns, becuase we removed empty rows in our initial pre process

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

now that we have 2806 movise, we will calculate the distance between each two vectro, we will calculate the angel between vectors,the smaller the angel the similar the movies.
we will use  cosin_similarity funciton from sklearn which is the opposite of distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

create function which take move and find five most similar movies to that move

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)

```

```python
recommend('Mr. Turner')
```
